Bar Plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
Languages=['Hindi','arabic','English','Japanese']
People=[100,20,30,40]
ax.bar(Languages,People)
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
Food=['Biryani','pasta','rice']
People=[200,34,78]
ax.pie(People,labels=Food,autopct='%1.1f%%')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

# Seaborn

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#set the theme
sns.set_theme()
```

```python
#visualise the data(relativeplot)
sns.relplot(data =tips,x='total_bill',y='tip',hue='smoker',col='time',style='smoker',size='size')
```

```python
#load the iris dataset
iris=sns.load_dataset('iris')
iris.head()
```

```python
#ScatterPlot
sns.scatterplot(data=iris,x='sepal_length',y='petal_length',hue='species')
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
#countplot
sns.countplot(x='class',data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

Bar Chart

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
#houseprice dataset
#this load boston dataset is in numpy array format we want in dataframe format
from sklearn.datasets import load_boston
house_boston=load_boston()
house=pd.DataFrame(house_boston.data,columns=house_boston.feature_names)
house['Price']=house_boston.target
```

```python
house.head()
```

Distribution Plot

```python
#want to find distribution of price
sns.distplot(house.Price)
```
